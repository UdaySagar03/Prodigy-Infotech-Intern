# Stacked Bar Plot

## Top 10 countries with male and female population 

```python
#merge male and female data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))

```

```python
#merged data
#calculate the toatal no of population for each country(male+female)
merge_data["Total_population"]=merge_data["2022_male"]+merge_data["2022_female"]
```

```python
merge_data.head()
```

```python
#sort data based on the total population in descending order
sorted_data = merge_data.sort_values(by="Total_population", ascending=False)
```

## Select the top 10 countries with the highest total population

```python
top_10_countries= sorted_data.head(10)
```

```python
#create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data=top_10_countries,color="red",label="female population")
sns.barplot(x="Country Code",y="2022_female",data=top_10_countries,bottom=top_10_countries["2022_female"],color="green",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

 ## Bottom 10 countries with the highest total population

```python
#select the bottom 10 countries with highest total population
bottom_10_countries= sorted_data.tail(10)
```

```python
#create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data=bottom_10_countries,color="red",label="female population")
sns.barplot(x="Country Code",y="2022_female",data=bottom_10_countries,bottom=bottom_10_countries["2022_female"],color="green",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
